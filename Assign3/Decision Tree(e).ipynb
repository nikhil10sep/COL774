```python
from read_data import read_data, read_data_cont
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
train = read_data_cont("dtree_data/train.csv")
valid = read_data_cont("dtree_data/valid.csv")
test = read_data_cont("dtree_data/test.csv")

bootstrap = False

# train = read_data("dtree_data/train.csv")
# valid = read_data("dtree_data/valid.csv")
# test = read_data("dtree_data/test.csv")
```

```python
file = open('forest_n_estimators_' + str(bootstrap).lower() + '.txt', 'w')
for n in range(1, 100, 2):
    clf = RandomForestClassifier(criterion='entropy', bootstrap=bootstrap, n_estimators=n, n_jobs=-1)
    clf.fit(train[:, 1:], train[:, 0])
    acc = [0] * 3
    acc[0] = str(clf.score(train[:, 1:], train[:, 0]))
    acc[1] = str(clf.score(valid[:, 1:], valid[:, 0]))
    acc[2] = str(clf.score(test[:, 1:], test[:, 0]))
    file.write(str(n) + ', ' + ', '.join(acc) + '\n')

file.close()
```

```python
file = open('forest_max_features_' + str(bootstrap).lower() + '.txt', 'w')
for k in range(1, 15):
    clf = RandomForestClassifier(criterion='entropy', bootstrap=bootstrap, max_features=k, n_jobs=-1)
    clf.fit(train[:, 1:], train[:, 0])
    acc = [0] * 3
    acc[0] = str(clf.score(train[:, 1:], train[:, 0]))
    acc[1] = str(clf.score(valid[:, 1:], valid[:, 0]))
    acc[2] = str(clf.score(test[:, 1:], test[:, 0]))
    file.write(str(k) + ', ' + ', '.join(acc) + '\n')

file.close()
```

```python
# Best parameters
clf = RandomForestClassifier(criterion='entropy', max_features=7, max_depth=30, min_samples_split=0.0007, min_samples_leaf=0.0001, n_estimators=30, n_jobs=-1)
clf.fit(train[:, 1:], train[:, 0])
print(clf.score(train[:, 1:], train[:, 0]))
print(clf.score(valid[:, 1:], valid[:, 0]))
print(clf.score(test[:, 1:], test[:, 0]))
```
