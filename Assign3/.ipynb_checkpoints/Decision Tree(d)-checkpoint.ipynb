```python
from read_data import read_data, read_data_cont
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
train = read_data_cont("dtree_data/train.csv")
valid = read_data_cont("dtree_data/valid.csv")
test = read_data_cont("dtree_data/test.csv")

# train = read_data("dtree_data/train.csv")
# valid = read_data("dtree_data/valid.csv")
# test = read_data("dtree_data/test.csv")
```

```python
file = open('dtree_depth.txt', 'w')
for depth in range(1, 30):
    clf = DecisionTreeClassifier(criterion='entropy', max_depth=depth)
    clf.fit(train[:, 1:], train[:, 0])
    acc = [0] * 3
    acc[0] = str(clf.score(train[:, 1:], train[:, 0]))
    acc[1] = str(clf.score(valid[:, 1:], valid[:, 0]))
    acc[2] = str(clf.score(test[:, 1:], test[:, 0]))
    file.write(str(depth) + ', ' + ', '.join(acc) + '\n')

file.close()
```

```python
vals = [1e-4, 3e-4, 7e-4, 1e-3, 3e-3, 7e-3, 1e-2, 3e-2, 7e-2, 1e-1]
file = open('dtree_min_samples.txt', 'w')
for frac in vals:
    clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=frac)
    clf.fit(train[:, 1:], train[:, 0])
    acc = [0] * 3
    acc[0] = str(clf.score(train[:, 1:], train[:, 0]))
    acc[1] = str(clf.score(valid[:, 1:], valid[:, 0]))
    acc[2] = str(clf.score(test[:, 1:], test[:, 0]))
    file.write(str(frac) + ', ' + ', '.join(acc) + '\n')

file.close()
```

```python
vals = [1e-5, 3e-5, 5e-5, 7e-5, 1e-4, 3e-4, 5e-4, 7e-4, 1e-3, 3e-3, 5e-3, 7e-3, 1e-2, 3e-2, 7e-2, 1e-1]
file = open('dtree_min_leaf.txt', 'w')
for frac in vals:
    clf = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=frac)
    clf.fit(train[:, 1:], train[:, 0])
    acc = [0] * 3
    acc[0] = str(clf.score(train[:, 1:], train[:, 0]))
    acc[1] = str(clf.score(valid[:, 1:], valid[:, 0]))
    acc[2] = str(clf.score(test[:, 1:], test[:, 0]))
    file.write(str(frac) + ', ' + ', '.join(acc) + '\n')

file.close()
```

```python
# Best parameters
clf = DecisionTreeClassifier(criterion='entropy', max_depth=12, min_samples_split=0.0014, min_samples_leaf=0.0001)
clf.fit(train[:, 1:], train[:, 0])
print(clf.score(train[:, 1:], train[:, 0]))
print(clf.score(valid[:, 1:], valid[:, 0]))
print(clf.score(test[:, 1:], test[:, 0]))
```
