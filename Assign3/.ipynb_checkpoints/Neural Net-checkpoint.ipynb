```python
import numpy as np
import math
from sklearn.linear_model import LogisticRegression
from visualization import plot_decision_boundary as plt_db
from svmutil import *

def read_data(filename, datatype=float):
    data = np.loadtxt(filename, dtype=datatype, delimiter = ',', ndmin=2)
    return data
```

```python
def get_accuracy(pred, labels):
    return len(np.nonzero(pred == labels)[0]) / pred.shape[0]
```

```python
def logistic_reg(X, labels, C_value=1):
    logreg = LogisticRegression(C=C_value)
    logreg.fit(X, labels.ravel())
    return logreg

def predict_logistic_reg(data, logreg):
    labels = logreg.predict(data)
    return labels.reshape((labels.shape[0], 1))
```

```python
def sigmoid(z):
    s = 1.0 / (1.0 + np.exp(-z))
    return s

def diff_sigmoid(z):
    s = sigmoid(z)
    return s * (1 - s)
```

```python
def relu(z):
    return np.maximum(z, 0)

def diff_relu(z):
    diff = np.zeros(z.shape)
    diff[z > 0] = 1
    return diff
```

```python
def initialize(n, L, size=0.01):
    params = []
    
    if not L:
        params.append([size * np.random.rand(n, 1), size * np.random.rand(1)])
        return params
    
    params.append([size * np.random.rand(n, L[0]), size * np.random.rand(L[0])])
    
    for i in range(len(L) - 1):
        params.append([size * np.random.rand(L[i], L[i + 1]), size * np.random.rand(L[i + 1])])
    
    params.append([size * np.random.rand(L[-1], 1), size * np.random.rand(1)])
    
    for param in params:
        param[0] -= size / 2
        param[1] -= size / 2
    return params
```

```python
def forward_prop(X, params, activation='sigmoid'):
    vals = [X]
    z = [X]
    activation_func = {'sigmoid' : sigmoid, 'relu' : relu}
    for i in range(len(params) - 1):
        vals.append(z[-1] @ params[i][0] + params[i][1])
        z.append(activation_func[activation](vals[-1]))
    
    vals.append(z[-1] @ params[-1][0] + params[-1][1])
    z.append(sigmoid(vals[-1]))
    return vals, z
```

```python
def back_prop(vals, z, labels, params, activation='sigmoid'):
    diff_activation_func = {'sigmoid' : diff_sigmoid, 'relu' : diff_relu}
    delta = [None] * (len(params) + 1)
    delta[-1] = (z[-1] - labels) * diff_sigmoid(vals[-1])
    for i in range(len(params) - 1, 0, -1):
        delta[i] = (delta[i + 1] @ params[i][0].T) * diff_activation_func[activation](vals[i])

    return delta
```

```python
def update_params(delta, z, params, eta):
    for i in range(len(params)):
        params[i][0] -= eta * z[i].T @ delta[i + 1]
        params[i][1] -= eta * np.sum(delta[i + 1], axis=0)
    
    return
```

```python
def train_neural_net(X, labels, L, activation='sigmoid', eta=0.1, batch_size=100, epsilon=0.3, max_iters=20000, var_eta=False,):
    m, n = X.shape
    params = initialize(n, L)
   
    error, prev_error = 0, 0
    period = 50
    change = [batch_size / 2] * period
    avg_change = batch_size / 2
    file = open('outp.txt', 'w')
    iters, break_flag = 0, False
    while True:
        for i in range(0, m, batch_size):
            iters += 1
            batch, bacth_labels = None, None
            if i + batch_size < m:
                batch = X[i : i + batch_size, :]
                batch_labels = labels[i : i + batch_size, :]
            else:
                batch = X[i : m, :]
                batch_labels = labels[i : m, :]
            
            vals, z = forward_prop(batch, params, activation=activation)
            delta = back_prop(vals, z, batch_labels, params, activation=activation)
            if var_eta:
                update_params(delta, z, params, eta / math.sqrt(iters))
            else:
                update_params(delta, z, params, eta)
            
            diff = z[-1] - batch_labels
            error = (0.5 * diff.T @ diff)[0][0]
            
#             avg_change = avg_change - (change[iters % period] - abs(prev_error - error)) / period
            change[iters % period] = abs(prev_error - error)
            avg_change = sum(change) / len(change)
    
            if iters % 1000 == 0:
                print(iters, error, avg_change)
#                 print(change)
#                 file.write(str(iters) + ', ' + str(error) + ', ' + str(avg_change) + '\n')
#                 file.write(str(change) + '\n')
#                 input()
                
            if avg_change < epsilon or iters >= max_iters:
                break_flag = True
                break
                
            prev_error = error
        
        if break_flag:
            break
    
    print(prev_error)
    return params           
```

```python
def predict_neural_net(data, params):
    vals, z = forward_prop(data, params)
    y = z[-1]
    labels = np.zeros(y.shape)
    labels[y >= 0.5] = 1
    return labels
```

```python
X = read_data('toy_trainX.csv')
y = read_data('toy_trainY.csv')
m, n = X.shape
print(m, n)
print(y.shape)
L = [10]
params = train_neural_net(X, y, L, eta=0.05, epsilon=1e-4, batch_size=m, max_iters=10000)
```

```python
X_test = read_data('toy_testX.csv')
y_test = read_data('toy_testY.csv')
pred = predict_neural_net(X, params)
print(get_accuracy(pred, y))
pred = predict_neural_net(X_test, params)
print(get_accuracy(pred, y_test))
plt_db(lambda x:predict_neural_net(x, params),X, y.ravel())
```

```python
logreg = logistic_reg(X, y, C_value=10)
pred = predict_logistic_reg(X, logreg)
print(get_accuracy(pred, y))
plt_db(lambda x:predict_logistic_reg(x, logreg),X_test, y_test.ravel())
```

```python
def train_svm(X, y, params=None):
    return svm_train((y.ravel()).tolist(), X.tolist(), params)

def predict_svm(data, model):
    p_labs, p_acc, p_vals = svm_predict([1] * data.shape[0], data.tolist(), model)
    return (np.array(p_labs)).reshape((len(p_labs), 1))
```

```python
train_data = read_data('mnist_data/MNIST_train.csv')
np.random.shuffle(train_data)
X = train_data[:, :-1] / 255
y = np.zeros((train_data.shape[0], 1))
y[train_data[:, -1] == 8] = 1
print(len(np.nonzero(y == 1)[0]))
m, n = X.shape
```

```python
test_data = read_data('mnist_data/MNIST_test.csv')
X_test = test_data[:, :-1] / 255
y_test = np.zeros((test_data.shape[0], 1))
y_test[test_data[:, -1] == 8] = 1
```

```python
model = train_svm(X, y, params='-t 0')
```

```python
pred = predict_svm(X_test, model)
print(get_accuracy(pred, y_test))
```

```python
pred = predict_svm(X, model)
print(get_accuracy(pred, y))
```
